# Why Parameter Discovery Matters

Every data collection workflow has parameters.

In API work, parameters may be query terms, date ranges, result limits, sorting
rules, pagination tokens, and authentication scopes.

In scraping, parameters are often less formal: CSS selectors, wait conditions,
scroll limits, URLs, request headers, or stopping rules.

These choices decide what can enter the dataset.

## Packages Used in This Notebook

This notebook mainly uses `requests` for one purpose: asking the MediaWiki API for its built-in help text.

The important point is not collection yet. The point is learning how to discover possible parameters before deciding what to collect.


```python
parameter_questions = [
    "Where is this parameter documented?",
    "What values are allowed?",
    "What is the default if we do not set it?",
    "Does changing it change the population of records?",
    "Does it affect ethics, legality, privacy, or server load?",
    "Will we record this choice in provenance?",
]

for question in parameter_questions:
    print("-", question)
```

::: {.callout-note}
The default value of a parameter is still a methodological choice. It is just
one that someone else made before we arrived.
:::

# Make an API Parameter Inventory

A parameter inventory is not required by Python. It is documentation for humans.

```python
api_parameter_inventory = [
    {
        "parameter": "action",
        "example_value": "query",
        "meaning": "which API operation to run",
        "why_we_set_it": "we want to read/query wiki data",
    },
    {
        "parameter": "list",
        "example_value": "search",
        "meaning": "which query module to use",
        "why_we_set_it": "we want search-result records",
    },
    {
        "parameter": "srsearch",
        "example_value": "digital services act",
        "meaning": "the search string",
        "why_we_set_it": "defines the substantive population of candidate pages",
    },
    {
        "parameter": "srlimit",
        "example_value": 50,
        "meaning": "number of results requested per response",
        "why_we_set_it": "sets the batch size for pagination",
    },
]

for row in api_parameter_inventory:
    print(row)
```

# Scraping Selectors Are Parameters Too

Scraping does not always have formal parameter names. But selectors behave like
parameters because they decide which parts of a page become data.

```python
selector_inventory = [
    {
        "selector": ".quote",
        "meaning": "every element with class='quote'",
        "role": "one repeated record container",
        "risk": "breaks if the site renames the CSS class",
    },
    {
        "selector": ".text",
        "meaning": "element with class='text' inside each quote block",
        "role": "quote text field",
        "risk": "captures only one visible field",
    },
    {
        "selector": "a[href]",
        "meaning": "all link elements with an href attribute",
        "role": "link extraction",
        "risk": "mixes content links and navigation links",
    },
]

for row in selector_inventory:
    print(row)
```

# Browser Automation Parameters

Browser automation adds another layer of choices: waiting, scrolling, clicking,
headless mode, and timeouts.

```python
browser_parameter_inventory = [
    {
        "parameter": "headless",
        "example_value": True,
        "meaning": "whether the browser runs invisibly",
        "methodological_issue": "headless and visible browsers can behave differently",
    },
    {
        "parameter": "wait_condition",
        "example_value": ".quote appears",
        "meaning": "when the script decides the page is ready",
        "methodological_issue": "waiting too little can miss content",
    },
    {
        "parameter": "scrolls",
        "example_value": 2,
        "meaning": "how many times the script scrolls",
        "methodological_issue": "scroll limits define visible content on infinite-scroll pages",
    },
]

for row in browser_parameter_inventory:
    print(row)
```
